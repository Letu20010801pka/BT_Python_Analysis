### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/u.user).

```python
data = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/u.user',sep='|')
```

```python
data.head()
```

### Step 3. Assign it to a variable called users.

```python
data.groupby('occupation').age.mean()
```

### Step 4. Discover what is the mean age per occupation

```python
#do tuoi trung binh moi nghe
data.groupby('occupation').age.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
# Male tren moi nghe nghiep
#Neu la Male => True , Nguoc lai
data['is_male'] = data.gender.apply(lambda x: True if x=='M' else False)
```

```python
data.is_male
```

```python
#Ti le Male tren moi nghe nghiep va sap xep
(data.groupby('occupation').is_male.sum() / data.groupby('occupation').gender.count()).sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
#Do tuoi min va max cho moi nghe nghiep
data.groupby('occupation').age.agg(['min','max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
#Do tuoi trung binh giua nghe nghiep va gioi tinh
data.groupby(['occupation', 'gender']).age.mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
#ti le women va men
#create data cho gender

gender_occup = data.groupby(['occupation', 'gender']).agg({'gender': 'count'})
```

```python
gender_occup
```

```python
#Create data va so luong tung nghe nghiep
occup_count = data.groupby(['occupation']).count() 
```

```python
occup_count
```

```python
#ti le 
occup_gender = gender_occup.div(occup_count, level='occupation')
```
